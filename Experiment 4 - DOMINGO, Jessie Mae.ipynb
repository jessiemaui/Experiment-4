Problem 1. ECE Board Problem Example.

```python
import pandas

student = pandas.read_csv('board2.csv')
Math = student[(student['Hometown'] == 'Visayas') & (student['Math']<70)]
Vis = Math[['Name','Gender','Track','Math']]

print(Vis.to_string(index=False))
```

Problem 1. Letter A.

```python
import pandas

student = pandas.read_csv('board2.csv')
Electronics = student[(student['Hometown'] == 'Luzon') & (student['Electronics']>70) & (student['Track'] == 'Instrumentation')]
Instru = Electronics[['Name','GEAS','Electronics']]

print(Instru.to_string(index=False))
```

Problem 1. Letter B.

```python
import pandas

student = pandas.read_csv('board2.csv')
student['Average'] = student[['Math','Electronics','GEAS','Communication']].mean(axis=1)
Constant = student[(student['Hometown'] == 'Mindanao') & (student['Gender'] == 'Female') & (student['Average']>=55)]
Mindy = Constant[['Name','Track','Electronics']]

print(Mindy.to_string(index=False))
```

Problem 2. Creating a Visualization.

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,5))
plt.bar(student['Track'],student['Average'], color='#FFC0CB')
plt.xlabel("Track")
plt.ylabel("Average Score")
plt.title("Correlation of the Chosen Track and Average of Students", fontsize=15)

print(plt.show())
```

```python
plt.figure(figsize=(10,5))
plt.bar(student['Gender'],student['Average'], color='#C68FE6')
plt.ylabel("Average Score")
plt.xlabel("Gender")
plt.title("The Correlation of the Gender and Average of Students", fontsize=15)

print(plt.show())
```

```python
plt.figure(figsize=(10,5))
plt.bar(student['Hometown'],student['Average'], color = '#FF0000')
plt.xlabel("Hometown")
plt.ylabel("Average Score")
plt.title("The Correlation of the Hometown and Average of Students", fontsize=15)

print(plt.show())
```
